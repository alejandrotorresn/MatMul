```python
import json
import pandas as pd
```

```python
import os

files = os.listdir('.')

types = ['avx2_', 'avx512_','cublas_', 'cublas_async_', 'cublas_tensor_', 'cuda_tiled_', 'cuda_naive_', 'mkl_', 'openmp_', 'sycl_', 'sycl_gpu_']
sizes = ['32', '48', '64', '80', '96', '112', '128', '256', '384', '512', '640', '768', '896', '1024', '2048', '3072', '4096', '5120', '6144', '7168', '8192']

perf_data = pd.DataFrame(index=sizes, columns=types)


for file in files:  
    for name in types:
        for size in sizes:
            if name in file and size in file:
                with open(file, 'r') as fp:
                    line_numbers = [5, 6, 8]
                    lines = []
                    for i, line in enumerate(fp):
                        if i in line_numbers:
                            line = (line.strip()).split(' ')[0]
                            lines.append(line.strip())
                        elif i > 9:
                            break
                    perf_data[name][size] = lines
```
